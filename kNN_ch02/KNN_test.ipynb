### Numpy.tile 矩阵扩展与复制

### Numpy.sum,axis矩阵求和

```python
#行求和
print '行求和:\n',b.sum(axis=1)
#列求和
print '列求和:\n',b.sum(axis=0)
#整个矩阵求和
print '整个矩阵求和:\n',b.sum()
```

### 数组计数存入字典并按照次数降序排序

```python
import operator
classCount={}
labels=['small','mid','large']
y=np.array([1,2,3,3,2,3])
for i in range(y.shape[0]):
    voteIlabel=labels[y[i]-1]
    # get 方法第二个参数给字典赋默认值
    print classCount.get(voteIlabel,0)
    classCount[voteIlabel]= classCount.get(voteIlabel,0)+1
print classCount
print sorted(classCount.iteritems(),key=operator.itemgetter(1),reverse=True)
```

### 读取文件的内容到矩阵中 

```python
import numpy as np
fr=open('datingTestSet2.txt')
arrayOfLines=fr.readlines()
numberOfLines=len(arrayOfLines)
returnMat=np.zeros((numberOfLines,3))
index=0
for line in arrayOfLines:
    line=line.strip()
    listFromLine=line.split('\t')
    returnMat[index,:]=listFromLine[0:3]
    index+=1
print returnMat
```
